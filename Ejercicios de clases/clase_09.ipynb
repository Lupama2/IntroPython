## Ejercicios 09 (a)

1. Realizar un programa para visualizar la función
   $$ f(x,n,w) = x^n |\sin(w x)| $$

   El programa debe realizar el gráfico para $w=10$, con cuatro curvas para $n=1,2,3,4$, similar al que se muestra
   en la siguiente figura

   ![](figuras/ejerc_vis_f1.png)

------

```python
#Defino f:
import numpy as np
def f(x,n,w):
    return x**n*np.abs(np.sin(w*x))
```

```python
#Grafico f
import matplotlib.pyplot as plt

w = 10
n_array = np.array([1,2,3,4])
for n in n_array:
    x_array = np.linspace(0,2*w)
    plt.plot(f(x_array,n,w))
```


Para usar latex podemos
(1) Escribir doble barra en un string común "\\mu" cosa de que no se confunda con los comandos del estilo \n
(2) Escribir al inicio del string una r, de la forma r"\mu". Este es un "rho-string" y significa que va tomar el string literalmente

También se puede llamar directamente a latex. Para hacer esto es necesario cambiar la configuración de matplotlib. Está bueno usar latex porque el paquete es mucho más grande que el que contiene matplotlib y por lo tanto, podemos hacer muchas más cosas

## Ejercicios 09 (b)

2. Para la función definida a trozos:
   $$
   f(x) =
   \begin{cases}
     f_{1}(x) = x^{2}/8 & - \pi < x \le \pi/2  \\
     f_{2}(x) = -0.3 x  & \pi/2 < x < \pi \\
     f_{3}(x) = -(x - 2 \pi)^{2}/6 & \pi \le x \le 5 \pi/2 \\
     f_{4}(x) = (x - 2 \pi)/5 & 5 \pi/2 < x \le 3 \pi
     \end{cases}
   $$
   realizar la siguiente figura de la manera más fiel posible.

   ![](figuras/ejercicio_08_1.png)

   **Pistas:** Buscar información sobre `plt.fill_between()` y sobre `plt.xticks` y `plt.yticks`. 

3. Rehacer la siguiente figura:

   ![](figuras/ejercicio_08_2.png)

-----

.

```python
import math
pi = math.pi

def f1(x):
    return x**2/8

def f2(x):
    return -0.3*x

def f3(x):
    return -(x-2*pi)**2/6

def f4(x):
    return (x-2*pi)/5

#Defino los colores y las labels
color_array = np.array(["b", "r", "k", "g"])
label_array = np.array([r"$f_1(x)$", r"$f_2(x)$", r"$f_3(x)$", r"$f_4(x)$"])

#Defino array de funciones
f_array = [f1,f2,f3,f4]

#Defino los arrays en x
N= 100
x1_array = np.linspace(-pi,pi/2,N)
x2_array = np.linspace(pi/2,pi,N)
x3_array = np.linspace(pi,5*pi/2,N)
x4_array = np.linspace(5*pi/2,3*pi,N)
x_array = [x1_array, x2_array, x3_array, x4_array]

#Grafico
indice = 0
x_min = -1.5

plt.plot(x_array[indice], f_array[indice](x_array[indice]), color = color_array[indice], label = label_array[indice])
plt.text(-1.5, 0.7, label_array[indice], color = color_array[indice])
plt.fill_between(x_array[indice], f_array[indice](x_array[indice]), np.zeros(N) + x_min, color = "b", alpha = 0.3)

indice += 1
plt.plot(x_array[indice], f_array[indice](x_array[indice]), color = color_array[indice], label = label_array[indice])

indice += 1
plt.plot(x_array[indice], f_array[indice](x_array[indice]), color = color_array[indice], label = label_array[indice])

indice += 1
plt.plot(x_array[indice], f_array[indice](x_array[indice]), color = color_array[indice], label = label_array[indice])

# plt.legend()
```
